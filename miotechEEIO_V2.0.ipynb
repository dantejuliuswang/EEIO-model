```python
# 计算出最终需求矩阵的对角阵, 注：这里已经把进口IM列减去了

Y_=[i.iloc[:153,163]-i.iloc[:153,164] for i in IOtable]

Y_[1]
```

```python
# 最终产出矩阵X，变成一维numpy数组
X=[i.iloc[:153,165].to_numpy() for i in IOtable]
X
```

```python
# 求出需求系数矩阵A
A=[(i.iloc[:153,:153]/i.iloc[159,0:153]).to_numpy() for i in IOtable]
A[0],A[1]
```

```python
n=153
I=np.identity(n)
I
```

```python
L=[np.linalg.inv(I-i) for i in A]
L[1]
```

```python
# 求出直接碳强度f,转换成对角阵
# f=IO_emissions_series/X
f_new_2018=pd.read_excel(r"D:\工作文件夹\妙盈\03. 酒店\map_project\total_emissions_factors_2018.xlsx")
f_new_2020=pd.read_excel(r"D:\工作文件夹\妙盈\03. 酒店\map_project\total_emissions_factors_2020.xlsx")
f_new=[f_new_2018["总排放因子"],f_new_2018["CO2排放因子"],f_new_2020["总排放因子"],f_new_2020["CO2排放因子"]]
f_new[3]
```

```python
# f_diag有4个elements
f_diag=[np.diag(i.to_numpy()) for i in f_new]  #对角阵的列表
f_diag[3]
```

```python
# 求出总排放因子,"整数除法//"
Total_factors=[f_diag[i] @ L[i // 2] for i in range(len(f_diag))]
# 消费碳排放因子=pd.DataFrame(Total_factors,index=level_1_index,columns=level_1_index)
# 消费碳排放因子=消费碳排放因子.sum(axis=0)
#消费碳排放因子.to_excel("消费碳排放因子_new.xlsx")
# np.shape(Total_factors[3])
Total_factors[3]
```

```python
Total_factors=[pd.DataFrame(i,index=level_1_index,columns=level_1_index) for i in Total_factors]
Total_factors[3]
```

```python
# 求出scope1排放因子，是行向求和还是列向求和？
factor1=[i@I.sum(axis=0) for i in f_diag]
factor1[1]
```

```python
# 求出scope2排放因子,即电力行业单独摘出
factor2=[(f_diag[i]@A[i//2]) for i in range(len(f_diag))]
factor2=[i[99,:] for i in factor2]
factor2[3]
```

```python
factor3=[Total_factors[i].sum(axis=0)-factor1[i]-factor2[i] for i in range(len(f_diag))]
factor3[0]
```

```python
#创建3个scope排放因子dataframe

# 将 ndarray 组成字典
data = [{'范围1排放因子': factor1[i], '范围2排放因子': factor2[i], '范围3排放因子': factor3[i]} for i in range(0,4)]
data[0]
```
